# Volatility Modeling Using Quantitative Methods

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import warnings
warnings.filterwarnings("ignore")
```

# TRADITIONAL RISK MODELS

### ARCH MODEL

```python
from arch import arch_model
tickers=['JPM']
df=yf.download(tickers, start='2022-01-01', end='2025-01-01')
df['returns']=df['Close'].pct_change()*100 # Daily % Return
returns=df['returns'].dropna()
```

```python
# Call the Model
model=arch_model(returns, vol='ARCH', p=1) 
# Train The Model
results=model.fit(disp='off')
results.summary()
```

Analysis of the ARCH Model
1. mu => 0.0978 : The model estimates that the average daily return is 0.1260
2. omega => 2.1877: The long run base level of variance
3. alpha[1] => 0.1224: How much yesterday's squared shock impacts today's variance

```python
# Forecast for next 5 days
forecast=results.forecast(horizon=5)
predicted_variance=forecast.variance
predicted_volatility=np.sqrt(predicted_variance)
predicted_volatility=[1.290791,	1.538572,	1.631696	,1.669358	,1.684967]
predicted_avg_volatility=sum(predicted_volatility)/len(predicted_volatility)
print(predicted_avg_volatility)
```

```python
# Actual Volatility
start_date=pd.to_datetime('2024-12-31')
end_date=pd.to_datetime('2025-01-09')

real_df=yf.download('JPM', start=start_date, end=end_date)
real_df['returns']=real_df['Close'].pct_change()*100
real_df=real_df.dropna()

realized_vol=real_df['returns'].std()*np.sqrt(5) # As we are predicting for 5 days
print(realized_vol)
```

```python
print("ARCH Model Predicted Volatility: ", predicted_avg_volatility)
print("ARCH Model Actual Volatility: ", realized_vol)
```

# GARCH Model

```python
from arch import arch_model
tickers=['JPM']
df=yf.download(tickers, start='2022-01-01', end='2025-01-01')
df['returns']=df['Close'].pct_change()*100 # Daily % Return
returns=df['returns'].dropna()
```

```python
# Call the Model
model=arch_model(returns, vol='GARCH', p=1, q=1) 

# Train The Model
results=model.fit(disp='off')

results.summary()
```

Analysis of the GARCH Model
1. mu => 0.1146 : THe model estimates that the average daily return is 0.1260
2. omega => 0.0220: The long run base level of variance
3. alpha[1] => 0.0165: How much yesterday's squared shock impacts today's variance
4. beta[1]=> 0.9737: How much yesterday's variance impacts today's variance

```python
# Forecast for next 5 days
forecast=results.forecast(horizon=5)
predicted_variance=forecast.variance
predicted_volatility=np.sqrt(predicted_variance)
predicted_volatility=[1.623482	,1.622251,	1.621031,	1.619823,	1.618625]
predicted_avg_volatility=sum(predicted_volatility)/len(predicted_volatility)
print(predicted_avg_volatility)
```

```python
# Actual Volatility
start_date=pd.to_datetime('2024-12-31')
end_date=pd.to_datetime('2025-01-09')

real_df=yf.download('JPM', start=start_date, end=end_date)
real_df['returns']=real_df['Close'].pct_change()*100
real_df=real_df.dropna()

realized_vol=real_df['returns'].std()*np.sqrt(5) # As we are predicting for 5 days
print(realized_vol)
```

```python
print("GARCH Model Predicted Volatility: ", predicted_avg_volatility)
print("GARCH Model Actual Volatility: ", realized_vol)
```

# EWMA Model

```python
# Download the data
tickers=['JPM']
df=yf.download(tickers, start='2024-01-01', end='2025-01-01')
df['returns']=df['Close'].pct_change()*100 # Daily % Return
df=df.dropna()
```

```python
# Calculate EWMA
lamda= 0.94
ewma_var=[]
var_t=df['returns'].var()

for ret in df['returns']:
    variance_tplus1=lamda*var_t+(1-lamda)*(ret**2)
    ewma_var.append(variance_tplus1)

df['ewma_vol']=np.sqrt(ewma_var)
```

```python
# Predicted Volatility (EWMA model can only be used for 1 day prediction)
latest_daily_vol=df['ewma_vol'].iloc[-1]
print(latest_daily_vol)
```

```python
start_date=pd.to_datetime('2024-12-31')
end_date=pd.to_datetime('2025-01-03')
real_df=yf.download('JPM', start=start_date, end=end_date)
real_df['returns']=real_df['Close'].pct_change()*100
real_df=real_df.dropna()

realized_vol=real_df['returns'] 
print(realized_vol)
```

```python
print("EWMA Model Predicted Volatility: ", latest_daily_vol)
print("EWMA Model Actual Volatility: ", realized_vol)
```
